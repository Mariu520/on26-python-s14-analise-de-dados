```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Seus dados
data = {
  
    'customer_zip_code_prefix': [14409, 9790, 1151, 8775, 13056],
    'customer_city': ['franca', 'sao bernardo do campo', 'sao paulo', 'mogi das cruzes', 'campinas'],
}

df = pd.DataFrame(data)

# Contagem das cidades
cidades = df['customer_city'].value_counts().head(5)

# Filtrando o DataFrame para considerar apenas as top 5 cidades
cidades_df = df[df['customer_city'].isin(cidades.index)]

# Criando o countplot usando Seaborn
plt.figure(figsize=(10, 6))
sns.countplot(y='customer_city', data=cidades_df, order=cidades.index, palette='viridis')
plt.title('As 5 Cidades de SP que Mais Solicitam Pedidos')
plt.xlabel('Número de Pedidos')
plt.ylabel('Cidade')
plt.show()
```

```python
import pandas as pd

# Carregar os dados do arquivo CSV
df_produtos = pd.read_csv('olist_order_items_dataset.csv')

# Agrupar por vendedor e somar o total de vendas
vendas_por_vendedor = df_produtos.groupby('seller_id')['price'].sum().reset_index()

# Ordenar os vendedores pelo total de vendas em ordem decrescente
vendedores = vendas_por_vendedor.nlargest(10, 'price')

# Exibir os resultados
print("Os 10 vendedores que mais venderam e o total vendido por cada um:")
print(vendedores)

```
